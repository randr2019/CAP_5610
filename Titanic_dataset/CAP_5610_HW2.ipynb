### Imputation based on KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
train=train_df.copy()
```

```python
test=test_df.copy()
```

```python
test['Fare'] = test['Fare'].fillna(test['Fare'].mode().iloc[0])
```

```python


def data_complete(data):
    knn_fet= data.filter (["Pclass","SibSp","Parch","Fare","Gender","Age"])
    data.drop(["Pclass","SibSp","Parch","Fare","Gender","Age"],1,inplace=True)
    imp_data = pd.concat([knn_fet,data],axis=1)
    print(imp_data.head())
    temp_train = imp_data.loc[imp_data.Age.notnull()] ## df with age values
    temp_test = imp_data.loc[imp_data.Age.isnull()] ## df without age values
    y = temp_train.Age ## setting target variables(age) in y 
    x = temp_train.loc[:, 'Pclass':'Gender']
    knn = KNeighborsRegressor(n_neighbors=5,weights='distance') 
    knn.fit(x, y)
    predicted_age = knn.predict(temp_test.loc[:, 'Pclass':'Gender'])
    imp_data.loc[imp_data.Age.isnull(), "Age"] = predicted_age
    
    return imp_data
```
